#### 2. Feature engineering

1. Feature scaling
2. Aggregation
3. One hot coding

```python
df.head(2)
```

```python
q25 = df["current_value"].quantile(0.25)
q75 = df["current_value"].quantile(0.75)
```

```python
q25, q75
```

```python

# Function to categorize prices
def categorize_price(value):
    if value < q25:
        return 'Low_Value'
    elif value < q75:
        return 'Mid_Value'
    else:
        return 'High_Value'

# Apply the function to create a new column
df['current_value_category'] = df['current_value'].apply(categorize_price)

df.drop('current_value', axis=1, inplace=True)

# Verify the distribution of the new categories
print(df['current_value_category'].value_counts())
```

```python
encoder = LabelEncoder()
df['current_value_category_encoded'] = encoder.fit_transform(df['current_value_category'])  
```

#### 4. Prepare train and test data

```python
#Prepare data
X = df


# sacle the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X.shape
```

## Buliding the Model

```python
wcss = []
for i in range(1, 20):  # Test 1 to 10 clusters
    kmeans = KMeans(n_clusters=i,
                    init='k-means++',
                    max_iter=300,
                    n_init=10,
                    random_state=0)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)  # inertia_ is the WCSS
```

```python
# 3. Plot the Elbow Curve
plt.figure(figsize=(10, 6))
plt.plot(range(1, 20), wcss,  marker='o', linestyle='-', color='b')
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.grid(True)
plt.show()
```

```python

optimal_clusters = 4 
model = KMeans(n_clusters=optimal_clusters, init='k-means++', max_iter=300, n_init=10, random_state=0)
```

## Train the Model

```python
y_pred = model.fit_predict(X_scaled)
```

## Test the Model

```python
# Predict and evaluate the model
y_pred = model.predict(X_scaled)
```

```python
y_pred
```

## Evaluating the Model 

```python
# Plotting the clusters with optimal number
plt.scatter(X_scaled[:, 0], X_scaled[:, 2], c=y_pred, s=50, cmap='viridis')

centers = model.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 2], c='red', s=200, alpha=0.75, marker='X')
plt.title(f'K-Means Clustering with {optimal_clusters} Clusters')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```
